```julia
using CSV
```

```julia
using Plots
```

```julia
using StatPlots
```

```julia
using DataFrames
```

```julia
dt = CSV.read("interpolation_time.csv", nullable=false)
```

```julia
data = by(dt, [:n, :type]) do subGroup
    DataFrame(mean_time = mean(subGroup[:time]), deviation = std(subGroup[:time]))
end
```

```julia
img = scatter(data[:n], 
    data[:mean_time], 
    colour = [:red :green :blue],
    group = data[:type],
    xlabel = "number of points",
    ylabel = "time [s]",
    yerr = data[:deviation])
```

```julia
Plots.pdf("graph.pdf")
```
