```julia
# mnożenie macierzy - wersja naiwna
function naive_multiplication(A,B)
C=zeros(Float64,size(A,1),size(B,2))
  for i=1:size(A,1)
    for j=1:size(B,2)
        for k=1:size(A,2)
            C[i,j]=C[i,j]+A[i,k]*B[k,j]
        end
    end
end
C
end
```

```julia
#kompilacja
naive_multiplication(Asmall,Bsmall)
```

```julia
#kompilacja funkcji BLASowej do mnożenia macierzy
#https://docs.julialang.org/en/stable/stdlib/linalg/#BLAS-Functions-1
Asmall*Bsmall
```

```julia
# poprawiona funkcja korzytająca z powyższego oraz z faktu, że
#można zmieniać kolejność operacji dodawania (a co za tym idzie kolejnosc petli).
function better_multiplication( A,B )
C=zeros(Float64,size(A,1),size(B,2))
  for j=1:size(B,2)
    for k=1:size(A,2)
        for i=1:size(A,1)
            C[i,j]=C[i,j]+A[i,k]*B[k,j]
        end
    end
end
C
end
```

```julia
using DataFrames
using CSV
```

```julia
function f(N)
    df1 = DataFrame(size = Int64[], version = String[], time = Float64[])
        for i = 1:N
                A = rand(Float64, 100*i, 100*i)
                B = rand(Float64, 100*i, 100*i)
                for j=1:10
                    naive_time = @elapsed naive_multiplication(A,B)
                    better_time = @elapsed better_multiplication(A,B)
                    blas_time = @elapsed A*B
                    df2 = DataFrame(size = 100*i, version = "naive", time = naive_time)
                    append!(df1, df2)
                    df2 = DataFrame(size = 100*i, version = "better", time = better_time)
                    append!(df1, df2)
                    df2 = DataFrame(size = 100*i, version = "blas", time = blas_time)
                    append!(df1, df2)
                    CSV.write("times_julia.csv", df1);
                end
        end
end
```

```julia
f(10)
```
