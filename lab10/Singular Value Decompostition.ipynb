1. Określenie słownika słów kluczowych (unia wszystkich słów występujących we wszystkich tekstach)

```python
import os
import re
import collections
import numpy as np
```

```python
allFiles = next(os.walk('.'))[2]
allFiles = list(filter(lambda x: re.search('.txt', x), allFiles))
allFiles
```

```python
texts = []
for text in allFiles:
    with open(text, 'r') as myfile:
        data=myfile.read().replace('\n', '')
    texts.append(data)
   
```

```python
bagsofwords0 = [collections.Counter(re.findall(r'\w+', txt)) for txt in texts]
bagsofwords = sum(bagsofwords0, collections.Counter())
bagsofwords
```

2. Wyznaczenie wektora cech bag-of-words dla każdego dokumentu. Zbudowanie rzadkiej macierzy wektorów bag-of-words, w której częstości cech ułożone są kolumnowo.
    

```python
bagsofwordsZeros = bagsofwords.copy()
for x in bagsofwordsZeros:
    bagsofwordsZeros[x] = 0
bagsofwordsZeros
```

```python
length = len(bagsofwords.values())
size = len(allFiles)
wordsMatrix = np.zeros((length, size))
words = set(bagsofwords.elements())
i = 0
j = 0

for text in allFiles:
    with open(text, 'r') as myfile:
        data=myfile.read().replace('\n', '')
        
    bagsofwordsVectorCollection = [collections.Counter(re.findall(r'\w+', data))]
    bagsofwordsVector = sum(bagsofwordsVectorCollection, collections.Counter())
    bagsofwordsZerosTmp = bagsofwordsZeros.copy()
    
    for word in set(bagsofwords):
        bagsofwordsVector[word] = bagsofwordsVector[word]
        wordsMatrix[i][j] = bagsofwordsVector[word]
        i+=1
    
    j+=1
    i=0
```

3. Redukcja znaczenia często występujących słów poprzez pomnożenie otrzymanego zbioru danych przez inverse document frequency.

```python
idfVector = np.zeros(length)
```

```python
i=0
for word in set(bagsofwords):
    idfVector[i] = np.log(size/bagsofwords[word])
    i+=1
idfVector
```

```python
i=0
j=0
for i in range(0, length):
    for j in range(0, size):
        wordsMatrix[i][j] = wordsMatrix[i][j]*idfVector[i]
```
