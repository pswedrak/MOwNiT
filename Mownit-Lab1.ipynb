<b> Zadanie 1 </b>  Porównać w języku Julia reprezentację bitową liczby 1/3 dla Float32, Float64 oraz liczby, 
która jest inicjalizowana jako Float32, a potem rzutowana na Float64.

```julia
x_32 = Float32(1/3)
println("Float32: ", bitstring(x_32))

x_64 = Float64(1/3)
println("Float64: ", bitstring(x_64))

x_32_to_64 = Float64(Float32(1/3))
println("Float32 to Float64: ", bitstring(x_32_to_64))
```

<b>Zadanie 4</b> Wymyślić własny przykład algorytmu niestablinego numerycznie. 
1. Zademostrować wersję niestabilną, pokazać, że działa źle.
2. Wyjaśnić, dlaczego działa źle.
3. Zademonstować wersję stabilną.
4. Język dowolny.

1. Obliczanie wyrazów ciągu: $$ x_{0} = 1, x_{1} = \frac{1}{4}, x_{n+1} = \frac{21}{4}x_{n} - \frac{5}{4}x_{n-1}$$ 

```julia
a = 1
b = 1/4
i = 1

while i < 1443
    c = 21/4 * b - 5/4 * a
    a = b
    b = c
    println(c)
    i = i+1
end
```

2. Algorytm jest niestabilny, ponieważ przy obliczaniu x_{n+1}, błąd przy x_{n} jest mnożony przez 21/4. -> wynik dąży do minus nieskończości

3. Wzór z punktu (1) tworzy ciąg o elementach $$x_{n} = (\frac{1}{4})^{n}$$

```julia
a = 1/4
i = 0
while i < 1443
    a = a/4
    println(a)
    i = i+1
end
```
